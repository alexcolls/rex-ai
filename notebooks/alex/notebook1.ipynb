```python


import pandas as pd

db_path = '../../db/data/tertiary/'

year = 2015

vols_ = pd.read_csv(db_path + str(year) +'/'+ 'vols_.csv', index_col=0)

vols_.index = pd.to_datetime(vols_.index)

y = vols_[:-25]

y
```

```python
from datetime import timedelta

# add time labels

ccys = y.columns

X = y

X['hour'] = X.index.hour
X['weekday'] = X.index.weekday
X['hour'] = X.index.hour
X['month'] = X.index.month
X['day'] = X.index.day
```

```python

for dt in X.index:
    if dt.hour == 12:
        for ccy in ccys:
            last_dt = dt - timedelta(days=1)
            print(dt, last_dt, X[ccy][dt])

```
